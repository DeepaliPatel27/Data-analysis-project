# Analyse Whether Daily change in closing price of stock or    Daily Returns in stock are co-reated or not!!!!

```python
pct_change = pd.DataFrame()

pct_change['AAPL_close_pct_change '] = app['close'].pct_change() *100
pct_change['AMZN_close_pct_change  '] = amazn['close'].pct_change() *100
pct_change['GOOG_close_pct_change  '] = goog['close'].pct_change() *100
pct_change['MSFT_close_pct_change ']= msft['close'].pct_change() *100
```

```python
pct_change
```

```python
h = sns.PairGrid(data = pct_change)
h.map_diag(sns.histplot)
h.map_lower(sns.scatterplot)
h.map_upper(sns.kdeplot)
```

```python
# another way of calculating pct change

closing_price['apple_close']
```

```python
closing_price['apple_close'].shift(1)  # shift(1) = ((68.56 - 67.85) / 67.85) *100
```

```python
(closing_price['apple_close'] - closing_price['apple_close'].shift(1)) / closing_price['apple_close'].shift(1) *100
```

```python
for col in closing_price.columns:
    closing_price[col +'_pct_change'] = (closing_price[col] - closing_price[col].shift(1)) / closing_price[col].shift(1) *100

```

```python
closing_price
```

```python
closing_price.columns
```

```python
clsing_p = closing_price[['apple_close_pct_change', 'amazn_close_pct_change',
       'goog_close_pct_change', 'msft_close_pct_change']]
```

```python
clsing_p
```

```python
g = sns.PairGrid(data = clsing_p)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```
